```python
track_lengths = {}

for tractor in a_tracks:
    track_lengths[tractor] = [((end - start)/60)/60 for start,end in a_tracks[tractor]]
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Get maximum subtrack length from all tractors
bar_width = 1
bigmax = 0
for tractor in track_lengths: bigmax = max(bigmax, max(track_lengths[tractor]))
bins = np.arange(0, bigmax+bar_width, bar_width)

# Get histogram values of subtrack lengths for all tractors
hist_list = {}
for tractor in track_lengths:
    hist_list[tractor] = np.histogram(track_lengths[tractor], bins=bins)[0]

# Make stacked histogram plot
plt.figure(figsize=(9,9))
bot = np.zeros_like(hist_list[tractor])
for tractor in hist_list:
    bot += plt.bar(bins[:-1]+bar_width/2, hist_list[tractor], width=bar_width,
                   bottom=bot, zorder=10, label=tractor).datavalues
plt.legend()
plt.title("Distribution of Track Durations")
plt.ylabel("Count")
plt.xlabel("Track Duration (hours)")

# Adjust axes
# plt.xticks(np.arange(0,bigmax+bar_width, 4))
plt.yscale("log")
plt.yticks(plt.yticks()[0], [f"{x:0.0f}" for x in plt.yticks()[0]])
plt.ylim(0.9, 1.1*bot[0])
plt.grid(True, "major", "y", zorder=1)
plt.grid(True, "minor", "y", linestyle=":", zorder=1)
```
